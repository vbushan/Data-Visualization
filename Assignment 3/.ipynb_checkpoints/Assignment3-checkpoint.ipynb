## Visual perception experiment

### Perception of length

Lets do a visual perception experiment where we guess the length of a horizontal bar by comparing it with two bars of know lengths. 

First, lets import the required libraries and define the lengths of the two bars.

```python
import random
import numpy as np
import time
```

```python
l_short_bar=1
mystery_length=random.uniform(1,10)
l_long_bar=10

plt.barh(np.arange(3), [l_short_bar, mystery_length, l_long_bar], align='center');
plt.yticks(np.arange(3), ('1', '?', '10'));
plt.xticks([]); # no hint!
```

Lets run the experiment once.

```python
perceived_length_list=[]
actual_length_list=[]

def run_exp_once():
    mystery_length = random.uniform(1, 10)  # generate a number between 1 and 10. 

    plt.barh(np.arange(3), [l_short_bar, mystery_length, l_long_bar], height=0.5, align='center')
    plt.yticks(np.arange(3), ('1', '?', '10'))
    plt.xticks([]) # no hint!
    plt.show()
    
    perceived_length_list.append( float(input("Guess the length of the middle bar ")) )
    actual_length_list.append(mystery_length)

run_exp_once()
```

## Task 3. Run the experiment at least TEN times and plot the result. 


```python
perceived_length_list=[]
actual_length_list=[]
def run_exp_ten():
    print('Guess the length of the middle bar')
    
    
    
    for i in range(1,11): # run 10 times
        mystery_length = random.uniform(1, 10)  # generate a number between 1 and 10. 

        plt.barh(np.arange(3), [l_short_bar, mystery_length, l_long_bar], height=0.5, align='center')
        plt.yticks(np.arange(3), ('1', '?', '10'))
        plt.xticks([]) # no hint!
        plt.title(f'Plot {i}')
        plt.show()

        perceived_length_list.append( float(input(f"Guess {i}- ")) )
        actual_length_list.append(mystery_length)
        

run_exp_ten()

# Start plotting the results
plt.scatter(x=perceived_length_list, y=actual_length_list)
x = perceived_length_list
y = actual_length_list
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x,p(x),"r--")

plt.show()
```

### Perception of area

Similar to the above experiment, now lets try to predict the area of a circle by comparing it with two circles of know areas.

```python
area1 = 0.005
area2 = 0.05

radius1 = np.sqrt(area1/np.pi) # area = pi * r * r
radius2 = np.sqrt(area2/np.pi)
random_radius = np.sqrt(area1*random.uniform(1,10)/np.pi)


plt.axis('equal')
plt.axis('off')
circ1 = plt.Circle( (0,0),radius1,clip_on=False,color='red', label="Area1"); # to change color to red add color = 'red'
circ2 = plt.Circle( (4*radius2,0), radius2, clip_on=False,color='green',label="Area2"); # to change color to green add color = 'green'
rand_circ = plt.Circle((2*radius2,0), random_radius, clip_on=False );

plt.gca().add_artist(circ1); # this adds axes to your circles, without that the circle won't draw
plt.gca().add_artist(circ2);
plt.gca().add_artist(rand_circ);
```

## Run the experiment once

```python
import math
perceived_area_list = []
actual_area_list = []
def run_area_exp_once(n1=0.005, n2=0.05):    
    radius1 = np.sqrt(n1/np.pi) # area = pi * r * r
    radius2 = np.sqrt(n2/np.pi)
    
    mystery_number = random.uniform(1,10)
    random_radius = np.sqrt(n1*mystery_number/math.pi)

    plt.axis('equal')
    plt.axis('off')
    circ1 = plt.Circle((0,0), radius1,clip_on=False,color='red')
    circ2 = plt.Circle((4*radius2,0),radius2,clip_on=False,color='green')
    rand_circ = plt.Circle((2*radius2,0), random_radius, clip_on=False )
    plt.gca().add_artist(circ1)
    plt.gca().add_artist(circ2)
    plt.gca().add_artist(rand_circ)    
    plt.show()
    
    perceived_area_list.append( float(input('Guess the area of the blue circle ')) )
    actual_area_list.append(mystery_number)
    
run_area_exp_once()
```

### Mix of different and same colored circles.

```python
perceived_area_list = []
actual_area_list = []
def run_area_exp_once(n1=0.005, n2=0.05):
    print('Guess the area of the blue circle')
    
    for i in range(1,11):
        
        radius1 = np.sqrt(n1/np.pi) # area = pi * r * r
        radius2 = np.sqrt(n2/np.pi)
    
        mystery_number = random.uniform(1,10)
        random_radius = np.sqrt(n1*mystery_number/math.pi)

        plt.axis('equal')
        plt.axis('off')
        
        if i>5:
            
            circ1 = plt.Circle((0,0), radius1,clip_on=False,color='red')
            circ2 = plt.Circle((4*radius2,0),radius2,clip_on=False,color='green')
        else:
            circ1 = plt.Circle((0,0), radius1,clip_on=False)
            circ2 = plt.Circle((4*radius2,0),radius2,clip_on=False)
            
            
        rand_circ = plt.Circle((2*radius2,0), random_radius, clip_on=False )
        plt.gca().add_artist(circ1)
        plt.gca().add_artist(circ2)
        plt.gca().add_artist(rand_circ)    
        plt.show()
    
        perceived_area_list.append( float(input(f'Guess {i}- ')) )
        actual_area_list.append(mystery_number)
    
run_area_exp_once()
```

```python
plt.scatter(x=perceived_area_list, y=actual_area_list)
x = perceived_area_list
y = actual_area_list
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x,p(x),"r--")

plt.show()
```

However, when I maintained different colors for each circle, my performance was better. I suspect that since both color(hue) and form(shape) are preattentive visual features, maintaing different color for each shape aided my estimation. 

Empirical results of research studies show that people in general correctly estimate lengths, but they struggle and understimate areas and volumes. Visualizations which include proportionally scaled circles have been the primary focus of many pyschophysical research studies which conclude that there is an inherent bias which leads to underestimating the areas and volumes. One research study suggests to use perceptual scaling to offset the underestimation. But, many scholars like, Tufte, have rejected the benefits of perceptual scaling. The concensus is that perceptual problem is not as evident in visualizations with a smaller range of circle sizes.

Visualizations which use proportionally scaled circles are quite common, and more recently the technique is used in many COVID trackers. One way they the handle perceptual problem is by using sequential palattes for indicating the density of cases in each area.
