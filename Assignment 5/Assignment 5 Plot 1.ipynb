```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv('Motor_Vehicle_Collisions_-_Crashes.CSV')
```

```python
df['CRASH YEAR'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y').dt.year
```

```python
data=df.groupby(['CRASH YEAR','BOROUGH']).count()['COLLISION_ID']
```

```python
BOROUGH=["BRONX","BROOKLYN","MANHATTAN","QUEENS","STATEN ISLAND"]
YEARS=list(range(2012,2021))
```

```python
plt.rcParams.update({'font.size': 18})

labels = YEARS
x = np.arange(len(labels))
width = 0.125

fig, ax = plt.subplots()
fig.set_size_inches(w=20,h=10)


w=0.125

k=-2
for i in range(len(BOROUGH)):
    
    temp=[]
    
    for j in range(len(YEARS)):
        temp.append(data[YEARS[j]][BOROUGH[i]])
    
    
    ax.bar(x+k*w*1.125 , temp, width, label=BOROUGH[i])
    k+=1
    


    
ax.set_ylabel('Total Collisions')
ax.set_title('Collisons by BOROUGH over Years ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
