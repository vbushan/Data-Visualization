```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv('NY_VC_DATA.CSV')
df=df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df=df.groupby(['CRASH YEAR']).sum()[['Pedestrians','Cyclists','Motorists']]
```

```python
df.head(len(df))
```

```python
groups=list(df.columns)
labels=list(df.index)
```

```python
plt.rcParams.update({'font.size': 18})

x = np.arange(len(labels))
width = 0.125

fig, ax = plt.subplots()
fig.set_size_inches(w=15,h=10)


w=0.125

k=-1

for i in range(len(groups)):
    
    temp=[]
    
    for j in range(len(labels)):
        temp.append(df.iloc[j][groups[i]])
    
    
    ax.bar(x+k*w*1.125 , temp, width, label=groups[i])
    k+=1
    


    
ax.set_title('Victims by type over Years ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.xlabel('Years')
plt.ylabel('Collisions')
fig.tight_layout()

plt.show()

```
