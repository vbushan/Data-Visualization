```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv('Motor_Vehicle_Collisions_-_Crashes.csv')
```

```python
df['CRASH YEAR'] = pd.to_datetime(df['CRASH DATE'], format='%m/%d/%Y').dt.year
df['CRASH HOUR']=pd.to_datetime(df['CRASH TIME'], format='%H:%M').dt.hour
```

```python
data=df.groupby(['BOROUGH','CRASH HOUR']).count()['COLLISION_ID']
```

```python
BOROUGH=["BRONX","BROOKLYN","MANHATTAN","QUEENS","STATEN ISLAND"]
```

```python
plt.rcParams.update({'font.size': 15})

fig, ax = plt.subplots()
fig.set_size_inches(w=20,h=10)

for i in range(len(BOROUGH)):
    data[BOROUGH[i]].plot(label=f"{BOROUGH[i]}")

plt.axvline(x=8,label='8am');
plt.axvline(x=16, label='4PM');
ax.annotate('8 AM',
            xy=(450, 40), xycoords='figure pixels')

ax.annotate('4 PM',
            xy=(800, 40), xycoords='figure pixels')
plt.title("Collisions by hour of day");
plt.legend();
plt.xlabel("Hour of day");
plt.ylabel('Collisions');
```
