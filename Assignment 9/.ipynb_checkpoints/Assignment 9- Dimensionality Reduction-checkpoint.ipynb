# PCA

Extract the numerical columns- 

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_only_features = iris[features]
iris_only_features.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # set the number of components to 2
```

```python
pca_iris_fitted = pca.fit(iris_only_features)
```

```python
pca_iris_fitted.explained_variance_ratio_
```

We can see that the two components explain nearly 93% of the variance in our dataset. So, we can be assured that the two prinicipal components are a decent approximation of all the dimensions in our dataset. 

Now, lets transform our dataset- 

```python
iris_pca = pca_iris_fitted.transform(iris_only_features)
iris_pca[:5]
```

Alternatively, we can do this- 

```python
iris_pca = pca.fit_transform(iris_only_features)
iris_pca[:5]
```

Now, let's convert the above matrix into a pandas dataframe- 

```python
iris_pca_df = pd.DataFrame(data=iris_pca, columns=['PC1', 'PC2'])
iris_pca_df.head()
```

# Task 5: Add a species column to iris_pca_df, create a pairplot based on the two principal component and discuss how this chart is different from what you generated above.

The first prinicipal component captures a significant amout of the variation in the dataset. Also, the datapoints from each class of species are clustered together - a preferred scenario for fitting a classification model.

# PCA with faces

```python
from sklearn.datasets import fetch_olivetti_faces

dataset = fetch_olivetti_faces(shuffle=True)
faces = dataset.data
```

```python
n_samples, n_features = faces.shape
print(n_samples)
print(n_features)
```

```python
print(faces[0].shape)
faces[0]
```

# Task 6. Reshape the one-dimensional array into an appropriate two-dimensional array to draw a face.

```python
face=faces[0].reshape((64,64))
plt.imshow(face);
```

Now, let's perform PCA on this dataset- 

```python
n_components=6
pca = PCA(n_components=n_components)
```

```python
pca.fit(faces)
```

```python
pca.components_
```

```python
pca.components_.shape
```

We can display the 6 components as images:

```python
image_shape = (64, 64)

for i, comp in enumerate(pca.components_, 1):
    plt.subplot(2, 3, i)
    plt.imshow(comp.reshape(image_shape), cmap=plt.cm.gray, interpolation='gaussian')
    plt.xticks(())
    plt.yticks(())
```

Now, lets transform the faces in our dataset- 

```python
faces_pca_transformed = pca.transform(faces)
```

```python
faces_pca_transformed.shape
```

```python
faces_pca_transformed[0]
```

It seems that the second component (with coordinate 4.14403343) contributes the most to the first face. Let's display them together and see how similar they are:

```python
# display the first face image 
plt.subplot(1, 2, 1)
plt.imshow(faces[0].reshape(image_shape), cmap=plt.cm.gray, interpolation='gaussian');
plt.xticks(())
plt.yticks(())

# display the second component
plt.subplot(1, 2, 2)
plt.imshow(pca.components_[1].reshape(image_shape), cmap=plt.cm.gray, interpolation='gaussian');
plt.xticks(());
plt.yticks(());
```

We can display the composition of faces in an "equation" style:

```python
from matplotlib import gridspec

def display_image(ax, image):
    ax.imshow(image, cmap=plt.cm.gray, interpolation='nearest')
    ax.set_xticks(())
    ax.set_yticks(())

def display_text(ax, text):
    ax.text(.5, .5, text, size=12)
    ax.axis('off')

face_idx = 0

plt.figure(figsize=(16,4))
gs = gridspec.GridSpec(2, 10, width_ratios=[5,1,1,5,1,1,5,1,1,5])

# display the face
ax = plt.subplot(gs[0])
display_image(ax, faces[face_idx].reshape(image_shape))

# display the equal sign
ax = plt.subplot(gs[1])
display_text(ax, r'$=$')

# display the 6 coordinates
for coord_i, gs_i in enumerate( [2,5,8,12,15,18] ):
    ax = plt.subplot(gs[gs_i])
    display_text( ax, r'$%.3f \times $' % faces_pca_transformed[face_idx][coord_i] )

# display the 6 components
for comp_i, gs_i in enumerate( [3,6,9,13,16,19] ):
    ax = plt.subplot(gs[gs_i])
    display_image( ax, pca.components_[comp_i].reshape(image_shape) )

# display the plus sign
for gs_i in [4,7,11,14,17]:
    ax = plt.subplot(gs[gs_i])
    display_text(ax, r'$+$')
```

```python
f, axes = plt.subplots(1, 6, figsize=(16,4))

faceid = 0

constructed_faces = []
for i in range(2, 10):
    constructed_faces.append(np.dot(faces_pca_transformed[faceid][:i], pca.components_[:i]))

# the face that we want to construct. 
display_image(axes[0], faces[0].reshape(image_shape))

for idx, ax in enumerate(axes[1:]):
    display_image(ax, constructed_faces[idx].reshape(image_shape))
```

# Non-negative Matrix Factorization (NMF)

Non-negative matrix factorization is a technique for obtaining low rank representation of matrices with non-negative or positive elements.By forcing every element in the decomposed matrices positive, NMF breaks something into parts that then can be added together.

```python
from sklearn.decomposition import NMF
import warnings
warnings.filterwarnings('ignore')

n_components=20
nmf = NMF(n_components=n_components)
nmf_fitted = nmf.fit(faces)

for i, comp in enumerate(nmf_fitted.components_, 1):
    plt.subplot(4, 5, i)
    plt.imshow(comp.reshape(image_shape), cmap=plt.cm.gray, interpolation='gaussian')
    plt.xticks(())
    plt.yticks(())
```

As you can see here, each 'component' of NMF picks up a certain part of the face (lighter area), such as eyes, chin, nose, and so on.

```python
faces_nmf_tranformed = nmf_fitted.transform(faces)

faces_nmf_tranformed[0]
```

# Task 7. Show the reconstructed faces using the first n components similar to how you did it for PCA above.

```python
f, axes = plt.subplots(1, 8, figsize=(16,4))

faceid = 0

constructed_faces = []
for i in range(2, 10):
    constructed_faces.append(np.dot(faces_nmf_tranformed[faceid][:i], nmf_fitted.components_[:i]))
    
    

# the face that we want to construct. 
display_image(axes[0], faces[0].reshape(image_shape))

for idx, ax in enumerate(axes[1:]):
    display_image(ax, constructed_faces[idx].reshape(image_shape))
```

# Task 9. Use imshow to display the first image. 

```python
plt.imshow(digits.data[0].reshape((8,8)));
```

Let's first reorder the data points according to the handwritten numbers.

```python
X = np.vstack([digits.data[digits.target==i]
               for i in range(10)])
y = np.hstack([digits.target[digits.target==i]
               for i in range(10)])
```

Now, let's use tSNE to reduce the dimensions in the dataset- 

```python
tsne = TSNE(n_components=2, init='pca', random_state=0)
```

```python
digits_proj = tsne.fit_transform(X)
```

```python
palette = np.array(sns.color_palette("hls", 10))
```

Now, let's plot the tranformed dataset- 

```python
plt.figure(figsize = (6,6))
plt.scatter(digits_proj[:,0], digits_proj[:,1],c=palette[y]);
```

We have isolated clusters. let's overlay the labels on each cluster- 

```python
def plot_scatter(projection):
    plt.figure(figsize = (6,6))
    plt.scatter(projection[:,0], projection[:,1],c=palette[y])
    for i in range(10):
        # Position of each label.
        xtext, ytext = np.median(projection[y == i, :], axis=0)
        txt = plt.text(xtext, ytext, str(i), fontsize=24)
```

```python
plot_scatter(digits_proj)
```

# Task 10. Create two plots with these two methods. 

```python
iso = Isomap( n_components=2).fit_transform(X)
plot_scatter(iso)
```

```python
mds = MDS(n_components=2).fit_transform(X)
plot_scatter(mds)
```

# Task 11. Write a paragraph comparing what you've learned about PCA, MDS, Isomap and t-SNE by visualizing the datasets.

<b>PCA: </b> PCA is a linear dimensionality reduction technique, and it does a decent job in approximating a dataset when the variation can be determined a set of orthogonal vectors/principal components. It only uses columns with numerical values, and cannot be used for a dataset with lot of categorical attributes. Also, orthogonal vectors do not always capture the variation in data.

<b> MDS/ISOMAP/t-SNE:</b> Other non-linear dimensional reduction techniques like MDS, Isomap and tSNE take some kind of distance metric (eucledian/geodesic) between points into count and optimize an objective function(RMSE/KL Divergence) such that the distance between data points with same labels is small in a lower dimensional space. All the methods are limited to numerical attributes.

Seeing the output plots, the tSNE method gave the best result. It preserves the local and global structures of the higher dimensional data.
